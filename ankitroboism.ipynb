```python
def list_string(list1,string):
    if string=="asc" :
        list1.sort()
        return list1
    elif string=="des" :
        list1.sort(reverse=True)
        return list1
    elif string=="None" :
        return list1
    else :
        print("wrong sorting option has been choosen")

n=int(input("no of values to be entered = "))
i=0
l=[]
print("enter the numerical values to be shorted ")
while(i<n) :
    l.append(int(input()))
    i=i+1

s=input("choose from asc,des,None  ")
print("the input list is :", l)
finallist= list_string(l,s)
print("this is the finallist after modification ==>7",finallist)
```

```python
def calculator(a, b, c):
    if(b=="+"):
        return int(a+c)
    if (b == "-"):
        return (a-c)
    if (b == "*"):
        return (a*c)
    if (b == "/"):
        return (a/c)

a=int(input("enter first number ==>"))
b=str(input("choose the operation to be performed from +,-,*,/ ==>"))
c=int(input("enter the second number ==>"))

d= calculator(a, b, c)
print(d)
```
